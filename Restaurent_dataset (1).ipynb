```python
price_del_book=df.groupby(by='Price range')[['Has Online delivery','Has Table booking']].value_counts().reset_index(name='Values')
price_del_book
```

```python
price_del=df.groupby(by='Price range')['Has Online delivery',].value_counts().reset_index(name='Values')
price_del
```

```python
import matplotlib.pyplot as plt
import numpy as np
price_range = [1, 2, 3, 4]
no_delivery = [3743, 1827, 997, 533]
yes_delivery = [701, 1286, 411, 53]
bar_width = 0.5
r = np.arange(len(price_range))
plt.bar(r, no_delivery, color='b', edgecolor='grey', width=bar_width, label='No Delivery')
plt.bar(r, yes_delivery, bottom=no_delivery, color='r', edgecolor='grey', width=bar_width, label='Yes Delivery')
plt.xlabel('Price Range', fontweight='bold')
plt.ylabel('Values', fontweight='bold')
plt.xticks(r, price_range)
plt.title('Stacked Bar Chart of Values by Price Range and Online Delivery')
plt.legend()
plt.show()
```

```python
price_tb=df.groupby(by='Price range')['Has Table booking'].value_counts().reset_index(name='Values')
price_tb
```

```python
price_range = [1,2,3,4]
no_tb= [4443,2874,764,312]
yes_tb=[1,239,644,274]
bar_width=0.5
s = np.arange(len(price_range))
plt.bar(s, no_tb, color='b', edgecolor='gray',width=bar_width, label='No Table booking')
plt.bar(s, yes_tb, color='r', edgecolor='gray',width=bar_width, label='Yes Table booking')
plt.xlabel('Price Range',fontweight='bold')
plt.ylabel('Values',fontweight='bold')
plt.xticks(s,price_range)
plt.title('Stack Bar Chart of Values by price Range and Online Delivery')
plt.legend()
plt.show()
```
